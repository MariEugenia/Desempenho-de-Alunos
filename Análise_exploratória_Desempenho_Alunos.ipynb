# Potenciais Problemas e Limitações e Comunicação de Resultados:

Houve muitos estudos sobre anotações, por exemplo, que implicam que elas aumentam significativamente o potencial acadêmico dos estudantes. Embora nossos dados sugiram uma correlação positiva com a tomada de notas, ela não é significativa. Acredito que isso ocorre devido à natureza auto-declarativa de uma pesquisa como essa, onde os estudantes podem ter diferentes interpretações do que significa 'às vezes'.

**Falta de Diversidade:**A tomada de notas é um tópico bastante polêmico entre os estudantes e, no entanto, só tem três opções disponíveis: 'Sempre', 'às vezes' e 'nunca', o que não captura a amplitude de um tópico potencialmente profundo como esse.

Baixa Correlação em Geral
Achei isso bastante surpreendente; quase todas as correlações presentes estão abaixo de 0,3, o que implica apenas uma correlação muito fraca. Existem dois valores discrepantes aqui: gênero e notas, e leitura. No entanto, os dados parecem indicar que não existem fatores determinantes fortes para o desempenho de um aluno que podem ser determinados por meio dessas informações demográficas. Em outras palavras, os resultados acadêmicos de um aluno não podem ser facilmente previstos com dados discretos como esses, o que implica que cada aluno tem a liberdade de determinar quais métodos são melhores para sua aprendizagem.
